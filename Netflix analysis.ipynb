```python
# show top 10 director who gave highest number of tv show and movies
director_count=data['Director'].value_counts().head(10)
director_count
```

```python
plt.figure(figsize=(8,5))
director_count.plot(kind='bar',color='brown')
plt.title("Top 10 Directors and Movies")
plt.ylabel('Movies')
plt.show()
```

```python
# the percentage between movies and tv show
category_count=data['Category'].value_counts()
category_count
```

```python
percentage=(category_count/category_count.sum())*100
percentage
```

```python
plt.pie(category_count,labels=category_count.index)
plt.legend()
plt.title("Movies By TV Show")
plt.show()
```

```python
# the best 10 countries with no. of movies and tv show
country_counts=data['Country'].value_counts()
country_data=country_counts.reset_index()
country_data.columns=['Country','Count']
country_data=country_data.sort_values(by='Count',ascending=False).head(10)
country_data
```

```python
plt.figure(figsize=(20,6))
plt.plot(country_data['Country'],country_data['Count'],marker='1',color='purple', linestyle='-' , linewidth=2)
# plt.xticks(rotation=90)
plt.title("Top 10 countries and its movie")
plt.xlabel('Countries')
plt.ylabel('No. Of Movies')
plt.grid(True)
plt.show()
```
